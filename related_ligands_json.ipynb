```python
import json

with open("related-ligands.json", "r") as f:
    data = json.load(f)

chem_comp_ids = [item["chem_comp_id"] for item in data.get("same_scaffold", [])]

# Print as comma-separated values
print(", ".join(chem_comp_ids))
```
